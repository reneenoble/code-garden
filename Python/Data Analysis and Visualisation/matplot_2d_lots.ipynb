## A Koala üê® Scatter Plot

```python
# Prepare tree/koala height data
import pandas as pd 

koala_data = pd.read_csv('data/koala_sightings_data.csv')
 
tree_heights = koala_data['HeightOfTree_m']
koala_heights = koala_data['HeightOfKoalaInTree_m']

print(koala_heights)
```

```python
# plot scatter plot

import matplotlib.pyplot as plt

plt.scatter(koala_heights, tree_heights)

plt.title('Koala tree position vs Tree height')
plt.xlabel('HeightOfKoalaInTree_m')
plt.ylabel('HeightOfTree_m')
```

# Make a bar chart üó∫üê®

```python
# Prepare the data, count the number of Koalas in each suburb

import pandas as pd 

koala_data = pd.read_csv('data/koala_sightings_data.csv') 
koala_sighting_count = koala_data.groupby('LGA')["_id"].count()

print(koala_sighting_count)
```

```python
# Make the bar chart
import matplotlib.pyplot  as plt

lgas = koala_sighting_count.keys()
counts = koala_sighting_count.values

# Make it a vertical bar chart
plt.bar(lgas, counts)

plt.title('Koala Sightings By LGA')
plt.xlabel('LGAs')
plt.ylabel('Koala Sightings')

plt.xticks(rotation = 45)

plt.savefig('koala_locations.png', bbox_inches='tight')
plt.show()
```

```python
# Make the bar chart
import matplotlib.pyplot  as plt

lgas = koala_sighting_count.keys()
counts = koala_sighting_count.values

# Make it a horizontal bar chart
plt.barh(lgas, counts)

plt.title('Koala Sightings By LGA')
plt.xlabel('LGAs')
plt.ylabel('Koala Sightings')

```

# Make a Line Chart üå°

```python
import csv
filename = "data/cairns-sst-2015.csv"

times = []
temps = []
with open(filename) as csvfile:
    # default is comma delimiter (but you can set one yourself)
    csv_reader = csv.reader(csvfile)
    header = next(csv_reader)
    for date_time, temp in csv_reader:
        times.append(date_time)
        temps.append(float(temp))
```

```python
import matplotlib.pyplot  as plt
from matplotlib.pyplot import figure



figure(figsize=(20, 6), dpi=80)
fig, ax = plt.subplots()

# times = times[0:100]
# temps = temps[0:100]

plt.plot(times, temps)
plt.xticks(rotation = 90)

labels = []
for i, time in enumerate(times):
    if i % 1000 == 0:
        labels.append(time)

    else:
        labels.append("")

ax.set_xticklabels(labels)
plt.margins(x=0)
```
